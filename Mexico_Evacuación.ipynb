```python
import pandas as pd

Mexico = pd.read_csv('DF_Mexico.csv', encoding='ISO-8859-1')
```

```python
#Identificar valores nulos por columna
valores_nulos=Mexico.isnull().sum()#.sum()
valores_nulos
```

# separar las columnas no numericas de las numericas

```python
#numericas 
#En este filtro estoy seleccionando unicamente la columna de fecha doc y la columna de nombre de vendedor 
#filtro3=df.iloc[ : , 1:3]       #Columnas consecutivas 

import pandas as pd

non_numeric_Mexico = Mexico.select_dtypes(exclude='number')
numeric_Mexico = Mexico.select_dtypes(include='number')
numeric_Mexico
```

#sustitucion de nulos

```python
# sustitucion de nulos por el promedio de la columnas 
numeric_Mexico=numeric_Mexico.fillna(round(Mexico.mean(),1))
numeric_Mexico
```

```python
# se eliminan las columnas innecesarias ya que solo tienen valores nulos 
numeric_Mexico = numeric_Mexico.drop(columns=['neighbourhood_group_cleansed', 'bathrooms',"calendar_updated","license"])
```

```python
#Identificar valores nulos por columna
valores_nulos=numeric_Mexico.isnull().sum()
valores_nulos
```

```python
#sustituir valores nulos por un string en concreto
non_numeric_Mexico=non_numeric_Mexico.fillna('sin registro')
```

```python
#Identificar valores nulos por columna
valores_nulos=non_numeric_Mexico.isnull().sum()
valores_nulos
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
numeric_Mexico.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

#**PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=numeric_Mexico
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers del Dataframe
outliers= numeric_Mexico[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data3= numeric_Mexico[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3
```

```python
#Eliminamos columnas innecesarias
#data4= data3.drop(['indice_tiempo'], axis=1)
#data4
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data3.copy()
data_clean=data_clean.fillna(round(data3.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
# Unimos la columna 'indice_tiempo' con el dataframe

#indice_tiempo=data2['indice_tiempo']
Datos_limpiosDSV= pd.concat([non_numeric_Mexico, data_clean], axis=1)
Datos_limpiosDSV.head(3)
```

```python
#Convertir DataFrame a CSV
Datos_limpiosDSV.to_csv("Datos_limpios_DSV_mexico.csv")
```

# **PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=numeric_Mexico

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= numeric_Mexico[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= numeric_Mexico[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
#Eliminamos columnas innecesarias
#data4_iqr= data3_iqr.drop(['indice_tiempo'], axis=1)
#data4_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data5_iqr=data3_iqr.copy()
print(data5_iqr.columns)
data5_iqr=data5_iqr.fillna(round(data3_iqr.mean(),1))
data5_iqr.columns
```

```python
# Unimos la columna 'indice_tiempo' con el dataframe

#indice_tiempo=data2['indice_tiempo']
Datos_limpiosIQR = pd.concat([non_numeric_Mexico, data5_iqr], axis=1)
Datos_limpiosIQR.head(3)
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Datos_limpiosIQR.isnull().sum().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
Datos_limpiosIQR.to_csv("Datos_limpios_IQR_mexico.csv")

```
